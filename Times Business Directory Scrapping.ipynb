```python
#Please install selenium using by removing the hash on line 2
#pip install selenium

#imports
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import csv
```

```python
#Amend the path to your chromedriver file accordingly 
#driver = webdriver.Chrome('/Users/kennethcheong/Downloads/chromedriver')
driver = webdriver.Firefox(executable_path='/Users/kennethcheong/Downloads/geckodriver')
```

```python
#Create list of urls to scrape from
main_url= 'https://www.timesbusinessdirectory.com/company-listings?page='
urls = []
for i in range(289,884):
    urls.append(main_url+str(i))
```

```python
xpath_1_1 = '//*[@id="Contentplaceholder1_C028_Col00"]/div[2]/div['
xpath_1_2 = ']/div/div[2]/p[1]/a'

email_list = []

for url in urls:
    driver.get(url)
    time.sleep(2)

    for i in range(2,12):
        try:
            driver.find_element_by_xpath(xpath_1_1+str(i)+xpath_1_2).click()
            time.sleep(2)

            try:
                driver.find_element_by_xpath('//*[@id="companyEmail"]').click()
                email_element = driver.find_element_by_id('companyEmail')
                email_list.append(email_element.text)
                print(len(email_list))
                driver.execute_script("window.history.go(-2)")

            except NoSuchElementException:
                driver.execute_script("window.history.go(-1)")
            
            except ElementNotInteractableException:
                driver.execute_script("window.history.go(-1)")
                
        except NoSuchElementException:
            continue
        except StaleElementReferenceException:
            continue
```

```python
df = pd.DataFrame(email_list)

# Remove duplicates
df.drop_duplicates(inplace=True, ignore_index=True)

# Export the dataframe to csv
df.to_csv('3.csv')
```
