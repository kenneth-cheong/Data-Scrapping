```python
#Please install selenium using by removing the hash on line 2
#pip install selenium

#imports
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
import pandas as pd
import csv
```

```python
#Amend the path to your chromedriver file accordingly 
driver = webdriver.Chrome('/Users/kennethcheong/Downloads/chromedriver')
```

```python
#Create list of urls to scrape from
main_url= 'http://www.siccmembers.com.sg/search/alphabetical?ViewCompany_page='
urls = []
for i in range(1,36):
    urls.append(main_url+str(i))
```

```python
#The scrapping magic begins
email_list = []

for url in urls:
    driver.get(url)
    
    email_element = driver.find_elements_by_class_name('email')
    
    for i in email_element:
        print(i.text)
        email_list.append(i.text)
```

```python
#Putting the collected emails into a Pandas DataFrame 
df = pd.DataFrame(email_list)
```

```python
# Remove 'Email\n:'
df = df.replace(to_replace='Email\n:',value='', regex=True)

# Remove duplicates
df.drop_duplicates(inplace=True, ignore_index=True)
```

```python
# Rxport the dataframe to csv
df.to_csv('all_countries_emails_only.csv')
```
